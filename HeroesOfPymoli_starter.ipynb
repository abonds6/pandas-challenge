## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#create the dataframe
popular_df = purchase_data_df.groupby("Item ID")["Item Name"].nunique()
#calculate the count of items
items = purchase_data_df.groupby("Item ID")["Item Name"].count()
#get the item names
item_name = purchase_data_df["Item Name"]
#calculate the count of purchases
purchase_count = purchase_data_df.groupby("Item ID")["Purchase ID"].count()
#calculate the average of purchases
item_price = purchase_data_df.groupby("Item ID")["Price"].mean()
#calculate the total of purchases
total_price = purchase_data_df.groupby("Item ID")["Price"].sum()
popular_df.head() 
#create the dataframe for summary
popular_summary = pd.DataFrame({
    "Item ID" : items,
    "Item Name" : item_name,
    "Purchase Count": purchase_count,
    "Item Price" : item_price,
    "Total Purchase Value" : total_price
})
#Sort by purchase count
popular_summary = popular_summary.sort_values(by="Purchase Count", ascending = False)
#format the fields
popular_summary["Item Price"] = popular_summary["Item Price"].map("${:.2f}".format) 
popular_summary["Total Purchase Value"] = popular_summary["Total Purchase Value"].map("${:.2f}".format) 
popular_summary  
```
